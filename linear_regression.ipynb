## Linear Regression (Using single feature)

```python
nonSmoker_df = df[df.smoker == 'no']
nonSmoker_df  #data formation
```

```python
plt.title('Age vs charges for non Smokers')

sns.scatterplot(data= nonSmoker_df,
                x='age',
                y='charges',
                alpha=0.7,
                s=15)
```

```python
def estimateCharges(age,w,b):
    return age*w+b;

# assuming that out linear regresssion forms a line with generalised formula y= wx + b where w is the slope (or weight) of the line and b is the constant  iintercept when the of x is 0 
    
```

```python
estimateCharges(30,50,100)
```

```python
#Assuming values of w and b
w=30;
b=45;
```

```python
ages =nonSmoker_df.age
estimateCharges = estimateCharges(ages,w,b)
estimateCharges
```

### line setting from down to upward
Lets try changing Paramters a littlebit,and lets make a function for the same

```python
def estimateCharges(age,w,b):
    return age*w+b;

def tryParams(w, b):
    ages = nonSmoker_df.age
    target = nonSmoker_df.charges

    estimatedCharges = estimateCharges(ages, w, b)  

    plt.plot(ages, estimatedCharges, 'r', alpha=0.9)
    plt.scatter(ages, target, s=30, alpha=0.8)
#s means size of dot


    plt.xlabel('Age')
    plt.ylabel('Charges')
    plt.legend(['Estimate', 'Actual'])
```

```python
tryParams(60, 300)
```

```python
tryParams(300,-4000)
```

## Now Instead of manually checking different values, wouldnt it be better if a computer could check various and give us the best result?

```python
#  finding the root mean square error 

import numpy as np
def rmse (targets,predictions):
    return np.sqrt(np.mean(np.square(targets-predictions)))
```

```python
w=50
b=100
```

```python
tryParams(w,b)
```

```python
targets = nonSmoker_df['charges']
predicted = estimateCharges(nonSmoker_df.age,w,b)
```

```python
rmse(targets, predicted)
```

```python
w=300
b=-4000

tryParams(w,b)
```

```python
targets = nonSmoker_df['charges']
predicted = estimateCharges(nonSmoker_df.age,w,b)
rmse(targets,predicted)
```
